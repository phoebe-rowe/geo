```python
import numpy as np
import pandas as pd
import geopy
from geopy.geocoders import Nominatim 
from geopy.extra.rate_limiterm import RateLimiter
```

```python
# get data 
url = "https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat"
data = pd.read_csv(url,header = None)
df = pd.DataFrame(data)
df.columns = ["Airport ID","Name","City","Country","IATA","ICAO","Latitude","Longitude",
              "Altitude","Time zone","DST","Tz","Type","Source"]

# pretend we don't know the address 
df = df.iloc[:,4:13]

# get column in correct latlong format
df['latlong'] = list(zip (df.Latitude, df.Longitude))
df.head()
```

```python
# define city fuction
def getcity(latlong):
    locator = Nominatim(user_agent="myGeocoder",timeout = 10)
    rgeocode = RateLimiter(locator.reverse,min_delay_seconds = 0.001)
    try:
        location = rgeocode(latlong)
        di = dict(location.raw)
        if 'city' in di.keys():
            city = di['city']
        elif 'village' in di.keys():
            city = di['village']
        elif 'town' in di.keys():
            city = di['town']
        else:
            city = 'no city or town'
    except:
        city = 'city not found'
    return city # or return location.raw to see all the data
```

```python
# apply function
df["city"] = df.iloc[:,9].apply(getcity)
df.head()
```
